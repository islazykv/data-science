# Predicting House Prices in King Country
The dataset was downloaded from kaggle ([link](https://www.kaggle.com/harlfoxem/housesalesprediction)).

The goal of this project is to create a model using Tensorflow 2.0 with Keras that predicts the final price of the households based on certain features.

#### List of the features:
* id - unique ID for each home sold
* date - date of the home sale
* price - price of each home sold
* bedrooms - number of bedrooms
* bathrooms - number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - square footage of the apartments interior living space
* sqft_lot - square footage of the land space
* floors - number of floors
* waterfront - a dummy variable for whether the apartment was overlooking the waterfront or not
* view - an index from 0 to 4 of how good the view of the property was
* condition - an index from 1 to 5 on the condition of the apartment,
* grade - an index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - the square footage of the interior housing space that is above ground level
* sqft_basement - the square footage of the interior housing space that is below ground level
* yr_built - the year the house was initially built
* yr_renovated - the year of the house’s last renovation
* zipcode - what zipcode area the house is in
* lat - lattitude
* long - longitude
* sqft_living15 - the square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - the square footage of the land lots of the nearest 15 neighbors

## Data PreProcessing
Let's check if some information can be extracted. Let's start with the 'date' feature.

###  Feature Engineering

```python
df['date'].head()
```

The date feature is written in the CUT (Coordinated Universal Time) time standard. It is be possible to convert it to a date object to extract the year and months, but instead I will slice it out as the format is fixed.

```python
df['year'] = df['date'].map(lambda x: x[0:4])
df['month'] = df['date'].map(lambda x: x[4:6])
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
round(df.groupby('year').mean()['price'][1]-df.groupby('year').mean()['price'][0])
```

```python
df.groupby('month').mean()['price'].plot()
```

As the data is from the years 2014 - 2015, on average the price increased by 2808 \\$. The best period for selling a house would be around spring time. Let's drop the date as it will not be neeeded anymore since we extracted important information.

```python
df.drop('date', axis=1, inplace=True)
```

Let's drop also the id and zipcode.

```python
df.drop(['id', 'zipcode'], axis=1, inplace=True)
```

## Data Preparation (Train Test Split)
Let's prepare the data for the model.

```python
y = df['price'].values
X = df.drop('price', axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

## Normalization (Scaling The Train Set)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.min()
```

```python
X_train.max()
```

## Building a Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linmodel = LinearRegression()
linmodel.fit(X_train,y_train)
```

```python
predictions = linmodel.predict(X_test)
```

```python
from sklearn import metrics

print('MAE:', np.round(metrics.mean_absolute_error(y_test, predictions),2))
print('MSE:', np.round(metrics.mean_squared_error(y_test, predictions),2))
print('RMSE:', np.round(np.sqrt(metrics.mean_squared_error(y_test, predictions)),2))
```

## Building a Neural Network Model
Let's build a model with 19 neurons (same number as features) with 5 hidden layers. Let's use ADAM as an optimizer.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,
          y=y_train,
          validation_data=(X_test,y_test),
          epochs=200,
          batch_size=128)
```

```python
df_loss = pd.DataFrame(model.history.history)
```

```python
df_loss.plot()
```

```python
predictions = model.predict(X_test)
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

<b> Mean Absolute Error (MAE) </b>

```python
mae = np.round(mean_absolute_error(y_test, predictions), 2)
mean = np.round(df['price'].mean(), 2)
mae_ratio = np.round(mae/mean*100, 2)
```

## Predicting a price of a random flat
Let's pick a random house and compare a real and predicted price of a house.

```python
import random
random.seed(101)
random_ind = random.randint(0,len(df))

random_house = df.iloc[random_ind]
real_price = random_house['price']

random_house = random_house.drop('price')
random_house = scaler.transform(random_house.values.reshape(-1, 19))
predicted_price = model.predict(random_house)

percent = np.round((real_price - predicted_price[0][0])/(real_price)*100, 2)
```

```python
print(f'''
Real price of the flat: {real_price} $,
Predicted price of the flat: {predicted_price[0][0]} $,
Difference in $: {real_price - predicted_price[0][0]} $
Difference in %: {percent} %
''')
```
